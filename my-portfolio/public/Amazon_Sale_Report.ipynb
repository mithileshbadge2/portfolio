```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d', errors='coerce')
```

```python
# monthly
if 'Date' in df.columns and 'Amount' in df.columns:
    df['Month'] = df['Date'].dt.to_period('M')
    monthly_sales = df.groupby('Month')['Amount'].sum()
    monthly_sales.plot(title='Monthly Sales Trends', figsize=(10, 6))
    plt.ylabel('Total Sales')
    plt.xlabel('Month')
    plt.show()
```

```python
# sales by category
if 'Category' in df.columns and 'Amount' in df.columns:
    category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)
    category_sales.plot(kind='bar', title='Sales by Category', figsize=(10, 6))
    plt.ylabel('Total Sales')
    plt.xlabel('Category')
    plt.show()
```

```python
categorical_cols = ['Courier Status', 'currency', 'ship-city', 'ship-state', 'ship-postal-code', 'ship-country', 'promotion-ids', 'fulfilled-by']
for col in categorical_cols:
    df[col] = df[col].fillna('Unknown')
```

```python
# Top-selling products (SKU)
if 'SKU' in df.columns and 'Amount' in df.columns:
    top_selling_skus = df.groupby('SKU')['Amount'].sum().sort_values(ascending=False).head(10)
    top_selling_skus.plot(kind='bar', title='Top 10 Selling Products (SKU)', figsize=(10, 6))
    plt.ylabel('Total Sales')
    plt.xlabel('SKU')
    plt.show()
```

```python
# Correlation Analysis
numerical_cols = ['Qty', 'Amount']
correlation_matrix = df[numerical_cols].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
